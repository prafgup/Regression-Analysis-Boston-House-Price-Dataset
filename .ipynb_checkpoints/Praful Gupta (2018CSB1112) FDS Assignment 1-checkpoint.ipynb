***Regression Analysis of Boston House Price Dataset***

```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
house_price = load_boston()
df = pd.DataFrame(house_price.data,
columns=house_price.feature_names)
df['PRICE'] = house_price.target
```

```python
df.head()
```

```python
df.keys()
```

```python
plot = sns.distplot(df["PRICE"],color="green",rug=True)
plot.set(xlabel='Price', ylabel='Distribution')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[house_price.feature_names], df["PRICE"], test_size=0.3, random_state=404)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(y_pred)
plt.plot(y_test.values)
```

```python
r2_s = r2_score(y_test.values,y_pred)
mse = mean_squared_error(y_test.values,y_pred)
```

```python
fig = plt.figure(figsize=(15, 6))
ax = fig.add_axes([0,0,1,1])
ax.bar(house_price.feature_names,model.coef_)
```

```python
required_features = house_price.feature_names
features_to_plot = ['RM','ZN','RAD','CRIM','TAX']
```

```python
def coeff_parser(coeff):
    co = []
    fet_ind = [ np.where(required_features == x) for x in features_to_plot ]
    for index,co in enumerate(coeff):
        coeff[index] = co[[5,1,8,0,9]]
    return coeff
```

```python

def singleRidgePipeline(alp,X_t,Y_t):
    r_model = Ridge(alpha=alp)
    r_model.fit(X_t,Y_t)
    return r_model.coef_

def multiRedge(X_t,Y_t,jump):
    coeff = []
    alp=0
    while alp<=200:
        coeff.append(singleRidgePipeline(alp,X_t,Y_t))
        alp += jump
    return coeff_parser(coeff)
```

```python
coeff_array_ridge = multiRedge(X_train[required_features],y_train,1)
plt.figure(figsize=(15, 10))
plt.ylim((-1,4))
plt.plot(coeff_array_ridge)
plt.legend(features_to_plot)
```

```python
def singleLassoPipeline(alp,X_t,Y_t):
    r_model = Lasso(alpha=alp)
    r_model.fit(X_t,Y_t)
    return r_model.coef_

def multiLasso(X_t,Y_t,jump):
    coeff = []
    alp=1
    while alp<=200:
        coeff.append(singleLassoPipeline(alp,X_t,Y_t))
        alp += jump
    return coeff_parser(coeff)
```

```python
coeff_array_lasso = multiLasso(X_train[required_features],y_train,1)
plt.figure(figsize=(15, 10))
plt.plot(coeff_array_lasso)
plt.legend(features_to_plot)
```

```python
def rid_lass3(a,X_t,Y_t):
    rid_model = []
    lass_model =[]
    for alp in a:
        r_model = Lasso(alpha=alp)
        r_model.fit(X_t,Y_t)
        lass_model.append(r_model)
        r_model = Ridge(alpha=alp)
        r_model.fit(X_t,Y_t)
        rid_model.append(r_model)
    return rid_model,lass_model
```

```python
ridge_models , lasso_models = rid_lass3([1,50,200],X_train[required_features],y_train)
```

```python
def potter(models,name):
    for i,mod in enumerate(models):
        plt.figure(3,figsize=(20, 10))
        plt.subplot(1,len(models),i+1)
        x = mod.predict(X_test)
        #print(len(x))
        plt.ylim(-15, 22)
        plt.xlim(-1, 42)
        plt.xlabel(name[i])
        plt.ylabel("Residuals")
        sns.residplot(x, y_test, lowess=True,color="g")
    
```

```python
potter(ridge_models,["Ridge Alpha = 1","Ridge Alpha = 50","Ridge Alpha = 200"])
```

```python
potter(lasso_models,["Lasso Alpha = 1","Lasso Alpha = 50","Lasso Alpha = 200"])
```

```python
potter([model],["Simple OLS Regression"])
```

```python
def train_test_error(mod):
    y_pred = mod.predict(X_test)
    test_r2_s = r2_score(y_test.values,y_pred)
    test_error_mse = mean_squared_error(y_test.values,y_pred)
    
    y_pred = mod.predict(X_train)
    train_r2_s = r2_score(y_train.values,y_pred)
    train_error_mse = mean_squared_error(y_train.values,y_pred)
    return train_error_mse,train_r2_s,test_error_mse,test_r2_s
    
    
```

```python
observation = pd.DataFrame(index = ["Linear Regression",
                                    "Ridge Regression alpha = 1",
                                    "Ridge Regression alpha = 50",
                                    "Ridge Regression alpha = 200",
                                    "Lasso Regression alpha = 1",
                                    "Lasso Regression alpha = 50",
                                    "Lasso Regression alpha = 200",],
                           columns=["Training Error (Mean Squared Error)",
                                    "R2 Score Training",
                                    "Test Error (Mean Squared Error)",
                                    "R2 Score Test"
                                   ],dtype=np.float32)
```

```python
observation.iloc[0][0],observation.iloc[0][1] ,observation.iloc[0][2] ,observation.iloc[0][3] = train_test_error(model)
observation.iloc[1][0],observation.iloc[1][1] ,observation.iloc[1][2] ,observation.iloc[1][3] = train_test_error(ridge_models[0])
observation.iloc[2][0],observation.iloc[2][1] ,observation.iloc[2][2] ,observation.iloc[2][3] = train_test_error(ridge_models[1])
observation.iloc[3][0],observation.iloc[3][1] ,observation.iloc[3][2] ,observation.iloc[3][3] = train_test_error(ridge_models[2])
observation.iloc[4][0],observation.iloc[4][1] ,observation.iloc[4][2] ,observation.iloc[4][3] = train_test_error(lasso_models[0])
observation.iloc[5][0],observation.iloc[5][1] ,observation.iloc[5][2] ,observation.iloc[5][3] = train_test_error(lasso_models[1])
observation.iloc[6][0],observation.iloc[6][1] ,observation.iloc[6][2] ,observation.iloc[6][3] = train_test_error(lasso_models[2])
```
